# Problem 1
Find the sum of all the multiples of
3 or 5
 below 1000
.

```python
def sum3or5(below):
  sum =0
  ans = ''
  for i in range(1,below):
    if ( i%3 ==0 or i%5 == 0):
      sum += i
      ans = ans  + str(i) + '+'
  return sum
```

```python
sum3or5(1000)
```

```python
palindrome = []
origin = []
def is_palindrome(number,text):

  # Handle negative numbers and 0
  if number < 0:
    return False
  elif number == 0:
    return True

  # Reverse the digits of the number
  reversed_number = 0
  original_number = number
  while number > 0:
    digit = number % 10
    reversed_number = reversed_number * 10 + digit
    number //= 10

  # Check if the reversed number is equal to the original number
  if reversed_number == original_number :
    palindrome.append(original_number)
    origin.append(text)
```

```python

for i in range (100,1000):
  for j in range (100,1000) :
    is_palindrome(i*j,f'{i}*{j}')
```

```python
print(max(palindrome))
```
