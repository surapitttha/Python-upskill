```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df  = pd.read_csv('/content/gdp-per-capita-worldbank.csv')
df.head()
```

```python
df['Entity'='thailand']
```

```python
filtered_df = df[df['Entity'] == 'Thailand']
```

```python
filtered_df.shape
```

```python
filtered_df.head()
```

```python
filtered_df.columns
```

```python
%matplotlib inline

plt.plot(filtered_df['Year'],filtered_df['GDP per capita, PPP (constant 2017 international $)'])
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

X = filtered_df['Year'].index.values.reshape(-1,1)
y = filtered_df['GDP per capita, PPP (constant 2017 international $)']


poly = PolynomialFeatures(degree=50)

X_poly = poly.fit_transform(X)

model = LinearRegression()

model.fit(X,y)
```

```python
y_pred  = model.predict(X)
y
```

```python

plt.plot(filtered_df['Year'],filtered_df['GDP per capita, PPP (constant 2017 international $)'],color='b')

plt.plot(filtered_df['Year'] ,y_pred,color='r')

plt.title("'GDP per capita, PPP (constant 2017 international $)'")
plt.xlabel('year')
plt.ylabel('GDP')
```
